## Linear Regression
---
##### Linear Regression Introduction
* Linear regression is a type of machine learning algorithm used to predict a continuous value i.e. house prices, revenue, crop yield, etc.
* For example:
    * Lot area to predict house prices
    * Advertisement cost to predict revenue
    * Amount of fertiliser used to predict crop yield
* These are examples of simple linear regression (SLR), where we are using only two features to predict a continuous value
    * The first feature is the constant value
    * The second feature is (i.e. lot area, advertisement cost, amount of fertiliser) 
* To make an accurate prediction, we often have to use multiple features. This is called Multiple Linear Regression (MLR)

##### Simple Linear Regression
* Here is the formula for SLR:

$h_\theta(x)^{(i)} = \theta_0 + \theta_1x_1^{(i)}$

* Where:
    * $h_\theta(x)$ (or $h(x)$ in short form) is our hypothesis function or predicted value
    * $i$ represents the ith training example
    * $j$ represents the jth feature
    * $m$ is the number of training examples
    * $n$ is the number of features. For a SLR, $n=2$, as there are always two features
    * $x_j$ represents the value of the jth input or feature. 
    * $\theta_j$ represents the jth parameter or coefficient 

##### Multiple Linear Regression
* Here is the formula for MLR:

$h_\theta(x)^{(i)} = \theta_0x_0^{(i)} + \theta_1x_1^{(i)} + ... + \theta_{j-1}x_{j-1}^{(i)}$

* Here is the vectorized form of MLR:

$h(X)= \theta_0x_1^{(i)} + \theta_1x_1^{(i)} + ... \theta_1x_{j-1}^{(i)}= X\theta =
\begin{pmatrix}
    X_{1,1} & X_{1,2} & \cdots & X_{1,n} \\
    X_{2,1} & X_{2,2} & \cdots & X_{2,n} \\
    \vdots  & \vdots  & \ddots & \vdots  \\
    X_{m,1} & X_{m,2} & \cdots & X_{m,n} 
\end{pmatrix}
\begin{pmatrix}
    \theta_{1}\\
    \theta_{2}\\
    \vdots \\
    \theta_{n} 
    \end{pmatrix}
= \begin{pmatrix}
    h_{1}\\
    h_{2}\\
    \vdots \\
    h_{m} 
    \end{pmatrix}$

* vectorization removes for loops and replacing them with vector multiplication, we can speed up the computation time
* Where:
    * $h(X)$ represents our predicted values and is a matrix of the size (m*1)
    * $X$ represents all the input values and is a matrix of the size (m*n)
    * $\theta$ represents all the parameters and is a matrix of the size (n*1)

```python
def preprocess_X(X):
    # This adds a column of ones in the first position to allow for matrix multiplication
    # x_1 is always 1 to denote the constant
    X = np.column_stack((np.ones(X.shape[0]),X))
    return X
```

### Cost Function
---
* We are trying to minimise the difference between our predicted value $h(x)$ with the actual target value $y$, this is called the cost function $J(\theta)$ or Mean Squared Error (MSE)
* Here is the cost function:

$J(\theta) = MSE = \frac{1}{2m}\sum_{i=1}^{m} (h(x)^{(i)} - y^{(i)})^2 = \frac{1}{2m}\sum_{i=1}^{m} (X\theta - y)^2= \frac{1}{2m} (X\theta - y)^T(X\theta - y)$

* By monitoring the cost function, we can tell if our algorithm is performing well or not

First we define the cost function, if we see the value decrease over iterations, we know our algorithm is working

```python
def cost(X,y,theta):
    m = X.shape[0]
    
    abs_diff = X@theta-y
    J=1/(2*m)*abs_diff.T@abs_diff
    
    return J
```

```python
# Setting the initial parameter
theta = np.array([0,0])
cost(X_slr,y_slr,theta)
```

## Gradient Descent
---

* To minimise the cost function, we need to take the partial derivative of the cost function in terms of each $\theta$:

$\frac{\partial J(\theta)}{\partial\theta} = \frac{1}{m}\sum_{i=1}^{m} (X\theta - y) X_j= \frac{1}{m} X^T(X\theta - y)
= \begin{pmatrix}
    \frac{\partial J(\theta)}{\partial\theta_{1}}\\
    \frac{\partial J(\theta)}{\partial\theta_{2}}\\
    \vdots \\
    \frac{\partial J(\theta)}{\partial\theta_{n}} 
    \end{pmatrix}$  

* Then simultaneously we update all the parameters:

$\theta := \theta - \alpha \frac{\partial J(\theta)}{\partial\theta}$  

* Where:
    * $\alpha$ is the learning rate (try 1, 0.3, 0.1, 0.03, 0.01, etc.)
    * Set it too low and it'll take forever to converge
    * Set it too high and the algorithm may never converge

```python
def update(X,y,theta,alpha,num_iter):
    m,n = X.shape 
    J_history = np.zeros(num_iter)
    
    for epoch in range(num_iter):
        
        abs_diff = X@theta-y
        grad = 1/m*X.T@abs_diff
        theta = theta - alpha*grad
    
        J_history[epoch]=cost(X,y,theta)
    
    fig_ud,ax_ud = plt.subplots(1,1,figsize=(16,5))
    ax_ud.plot(range(num_iter),J_history)
    ax_ud.set_xlabel('epoch',size=16);
    ax_ud.set_ylabel('Loss',size=16);
    ax_ud.set_title('Loss per epoch',size=20)
    return theta, J_history, fig_ud, ax_ud
```

## SLR Result

```python
# Takes about 1200 iterations to converge
new_theta_SLR,J_history,_,_=update(X_slr,y_slr,theta,0.01,1500)
```

## Plotting Cost
---
* The size of input values will heavily skew the shape of the contour loss curve
* A large input value will need less of a coefficient to generate huge losses
* While a small input value will need a huge coefficient to match the same amount of loss

```python
def plot_gradient_descent_SLR(X_slr,y_slr,actual_theta, theta_num):
    fig_gd,ax_gd = plt.subplots(1,1,figsize=(16,5))
    if theta_num == 0:
        theta = np.linspace(actual_theta[0]-5,actual_theta[0]+5,1000)
        cost_theta = [cost(X_slr,y_slr,np.array([i,actual_theta[1]])) for i in theta]
    elif theta_num ==1:
        theta = np.linspace(actual_theta[1]-5,actual_theta[1]+5,1000)
        cost_theta = [cost(X_slr,y_slr,np.array([actual_theta[0],i])) for i in theta]

    ax_gd.plot(theta,cost_theta)
    ax_gd.set_xlabel(f'Theta {theta_num}',size=16);
    ax_gd.set_ylabel('Cost',size=16);
    ax_gd.set_title('Gradient Descent',size=20);   
    return fig_gd, ax_gd
```

```python
# Note the maximum cost of theta_0 between -10 and 10 is ~16
plot_gradient_descent_SLR(X_slr,y_slr,new_theta_SLR,theta_num=0);
```

```python
# Note the maximum cost of theta_1 between -10 and 10 is ~1000
plot_gradient_descent_SLR(X_slr,y_slr,new_theta_SLR,theta_num=1);
```

```python
def plot_loss_contour_SLR(X_slr,y_slr,actual_theta):
    fig_lc,ax_lc = plt.subplots(1,1,figsize=(8,8))
    
    # Compute the cost for all theta_0 and theta_1 combinations
    theta_0 = np.linspace(actual_theta[0]-10,actual_theta[0]+10,100)
    theta_1 = np.linspace(actual_theta[1]-10,actual_theta[1]+10,100)
    cost_theta = np.zeros([len(theta_0),len(theta_1)])
    for i,theta_0_val in enumerate(theta_0):
        for j,theta_1_val in enumerate(theta_1):
            cost_theta[i,j] = cost(X_slr,y_slr,np.array([theta_0_val,theta_1_val]))
    
    # Construct two matrices that represents all the co-ordinates
    # Set the indexing to matrix indexing [row,col] and not cartesian indexing (col,row)
    X, Y = np.meshgrid(theta_0, theta_1,indexing='ij')
      
    # Plot the losses at log space levels
    CS = ax_lc.contour(X,Y,cost_theta,levels=(np.logspace(-2,3,20)))
    ax_lc.set_xlabel('Theta 0',size=16);
    ax_lc.set_ylabel('Theta 1',size=16);
    ax_lc.set_title('Loss contour',size=20);
    ax_lc.plot(actual_theta[0],actual_theta[1],'rx',linewidth='15');
    ax_lc.clabel(CS, inline=True, fontsize=10)
    
    final_cost = cost(X_slr,y_slr,actual_theta)
    ax_lc.text(actual_theta[0]+12,actual_theta[1]+5,f'theta: ({actual_theta[0]:.1f},{actual_theta[1]:.1f})\nloss: {final_cost:.2f}',size=20);
    return fig_lc,ax_lc
```

## Standard Scaling
---
* If we scale the data, we can get to the minimum quicker
* It will also remove the skew of our contour loss curve

```python
def scaler(X):
    # Exclude the first column
    X[:,1:]=(X[:,1:]-np.mean(X[:,1:],axis=0))/np.std(X[:,1:],axis=0)
    return X
```

```python
X_slr_ss=scaler(copy.copy(X_slr))
X_mlr_ss=scaler(copy.copy(X_mlr))
```

```python
# Takes about 400 iterations to converge
new_theta_SLR_ss,J_history,_,_=update(X_slr_ss,y_slr,theta,0.01,1500)
```
